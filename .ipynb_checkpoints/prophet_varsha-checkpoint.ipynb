```python
train = pd.read_csv("train.csv", index_col=1, parse_dates=True)
train=train.drop("id",axis=1)
test = pd.read_csv("test.csv", index_col=1, parse_dates=True)
test=test.drop("id",axis=1)
test["Close"]=close_forecast["yhat"].values
```

```python
t1=train
t1_dummies = pd.get_dummies(t1,columns=["Strategy"]).astype(float)
X = t1_dummies[["Close","Open","Volume"]]
Y = t1_dummies[["Strategy_Buy","Strategy_Hold","Strategy_Sell"]]
x_pred = test[["Close","Open","Volume"]]
```

```python
mapping={"Buy":0,"Hold":1,"Sell":2}
Y = train.Strategy.map(mapping)
Y
```

```python
X_train = X[["Close","Open","Volume"]][:240]
# y_train = Y[["Strategy_Buy","Strategy_Hold","Strategy_Sell"]][:240]
y_train=Y[:240]
X_test = X[["Close","Open","Volume"]][240:]
# y_test = Y[["Strategy_Buy","Strategy_Hold","Strategy_Sell"]][240:]
y_test=Y[240:]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the parameter grid for hyperparameter tuning
params = {
    'C': [0.001, 0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 200, 300]
}

best_score = 0  # Initialize best_score
best_params = {}  # Initialize best_params

# Iterate through the parameter grid
for C in params['C']:
    for penalty in params['penalty']:
        for solver in params['solver']:
            for max_iter in params['max_iter']:
                # Create a Logistic Regression model with the current hyperparameters
                clf = LogisticRegression(
                    C=C,
                    penalty=penalty,
                    solver=solver,
                    max_iter=max_iter,
                    random_state=42,
                    multi_class='ovr'  # 'ovr' for one-vs-rest (for multi-class classification)
                )

                # Train the model on the training train
                clf.fit(X_train, y_train)

                # Make predictions on the test train
                y_pred = clf.predict(X_test)

                # Calculate the accuracy score
                accuracy = accuracy_score(y_test, y_pred)

                # Check if the current model has a higher accuracy score
                if accuracy > best_score:
                    best_score = accuracy
                    best_params = {
                        'C': C,
                        'penalty': penalty,
                        'solver': solver,
                        'max_iter': max_iter
                    }

# Print the best accuracy score and corresponding hyperparameters
print("Best Accuracy Score:", best_score)
print("Best Hyperparameters:", best_params)
```

```python
x_pred
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train a decision tree classifier
classifier = DecisionTreeClassifier(max_depth=20,
                                    min_samples_split=2,  # Minimum samples required to split an internal node
                                    min_samples_leaf=1,
                                    criterion='gini',
                                    random_state=42)
classifier.fit(X,Y)
y_pred = classifier.predict(x_pred)
ans=[]
for i in range(len(y_pred)):
    if y_pred[i][0]==1:
        ans.append("Buy")
    elif y_pred[i][1]==1:
        ans.append("Hold")
    else:
        ans.append("Sell")
ans.count("Buy"), ans.count("Hold"), ans.count("Sell")
```

```python
submission = pd.trainFrame()
submission["Date"] = test.index
submission["Close"] = test["Close"].values
submission["Strategy"] = ans
submission.index.name="id"
submission.head()
```

```python
submission.to_csv('prophet.csv', index=True)
```
