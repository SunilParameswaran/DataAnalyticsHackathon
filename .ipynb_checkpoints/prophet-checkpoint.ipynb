# METHOD 2

```python
import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics

# Load the train.csv dataset
train_data = pd.read_csv('train.csv')
# Rename columns to fit Prophet's requirements
train_data = train_data.rename(columns={'Date': 'ds', 'Close': 'y', 'Open': 'open', 'Volume': 'volume'})
```

```python

# Create and fit the Prophet model with Open and Volume as regressors
model = Prophet()
model.add_regressor('open')
model.add_regressor('volume')

model.fit(train_data.head(300))  # Using the first 300 rows for training

# Define a parameter grid for hyperparameter tuning
param_grid = {
    'seasonality_prior_scale': [0.01, 0.1, 1.0],
    'holidays_prior_scale': [0.01, 0.1, 1.0],
    'changepoint_prior_scale': [0.001, 0.01, 0.1],
}

best_params = None
best_mse = float('inf')

# Perform hyperparameter tuning using cross-validation
for seasonality_prior in param_grid['seasonality_prior_scale']:
    for holidays_prior in param_grid['holidays_prior_scale']:
        for changepoint_prior in param_grid['changepoint_prior_scale']:
            model.seasonality_prior_scale = seasonality_prior
            model.holidays_prior_scale = holidays_prior
            model.changepoint_prior_scale = changepoint_prior

            # Cross-validation to evaluate the model with different hyperparameters
            df_cv = cross_validation(model, initial='300 days', period='30 days', horizon='30 days')
            df_p = performance_metrics(df_cv)
            mse = df_p['mse'].values[0]

            if mse < best_mse:
                best_mse = mse
                best_params = {
                    'seasonality_prior_scale': seasonality_prior,
                    'holidays_prior_scale': holidays_prior,
                    'changepoint_prior_scale': changepoint_prior,
                }

# Use the best hyperparameters to train the final model
model = Prophet(
    seasonality_prior_scale=best_params['seasonality_prior_scale'],
    holidays_prior_scale=best_params['holidays_prior_scale'],
    changepoint_prior_scale=best_params['changepoint_prior_scale']
)
model.add_regressor('open')
model.add_regressor('volume')

model.fit(train_data.head(300))  # Using the first 300 rows for training

# Load the test.csv dataset
test_data = pd.read_csv('test.csv')
test_data = test_data.rename(columns={'Date': 'ds', 'Open': 'open', 'Volume': 'volume'})

# Make predictions for the test data
forecast = model.predict(test_data)

# Extract the Close price forecasts from the predictions
close_forecast = forecast[['ds', 'yhat']]
close_forecast.index = [i for i in range(300, 400)]

# Print the Close price forecasts
print("Close Price Forecast:")
print(close_forecast)
```
