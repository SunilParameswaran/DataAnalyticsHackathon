```python
train = pd.read_csv("train.csv", index_col=1, parse_dates=True)
test = pd.read_csv("test.csv", index_col=1, parse_dates=True)
```

```python
decomposition = sm.tsa.seasonal_decompose(train['Close'], model='multiplicative', period=39)
plt.figure(figsize=(30, 10))
plt.subplot(4, 1, 1)
plt.plot(train['Close'], label='Original Groundwater Level')
plt.legend()
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Trend Component')
plt.legend()
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Seasonal Component')
plt.legend()
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Residuals')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Test for staionarity
def test_stationarity(timeseries):
    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
    
test_stationarity(train["Close"])
```

```python
model = SARIMAX(train["Close"],
                order=(2,1,0),
                seasonal_order=(1,1,0,52),
                exog=train[["Open","Volume"]],
                trend="n")
fitted = model.fit(disp=0)
```

```python
preds = fitted.forecast(100,exog=test[["Open","Volume"]])
# preds = np.exp(preds)
```

```python
t1 = train
t2 = test
```

```python
t1_dummies = pd.get_dummies(t1,columns=["Strategy"]).astype(float)
t1_dummies
```

```python
X = t1_dummies[["Open","Close","Volume"]]
Y = t1_dummies[["Strategy_Buy","Strategy_Hold","Strategy_Sell"]]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train a decision tree classifier
classifier = DecisionTreeClassifier(max_depth=15,
                                    min_samples_split=2,  # Minimum samples required to split an internal node
                                    min_samples_leaf=1,
                                    criterion='gini',
                                    random_state=42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(t2[["Open","Close","Volume"]])
```

```python
ans=[]
for i in range(len(y_pred)):
    if y_pred[i][0]==1:
        ans.append("Buy")
    elif y_pred[i][1]==1:
        ans.append("Hold")
    else:
        ans.append("Sell")
```

```python
ans.count("Buy"), ans.count("Hold"), ans.count("Sell")
```

```python
submission = pd.DataFrame()
submission["Date"] = preds.index
submission["Close"] = preds.values
submission["Strategy"] = ans
submission.index.name="id"
submission.head()
```

```python
submission.to_csv('sarima_seasonality.csv', index=True)
```
