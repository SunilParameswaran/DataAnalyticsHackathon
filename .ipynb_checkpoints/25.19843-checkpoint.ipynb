#### Decomposition into trend and seasonal components

```python
decomposition = sm.tsa.seasonal_decompose(train['Close'], model='additive')
plt.figure(figsize=(30, 10))
plt.subplot(4, 1, 1)
plt.plot(train['Close'], label='Original Groundwater Level')
plt.legend()
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Trend Component')
plt.legend()
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Seasonal Component')
plt.legend()
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Residuals')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(decomposition.seasonal, label='Seasonal Component')
```

#### Task 1

#### As we can see from the above graph - there is a seasonal component which repeats approximately every 39 weeks.

#### We use SARIMAX MODEL because our dataset contains seasonal cycles as seen from the above graphs

```python
train_close_differenced_values = train['Close'].diff().dropna()
print(train_close_differenced_values.head())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
# Plot the ACF (Autocorrelation Function)
plt.figure(figsize=(12, 6))
plot_acf(train_close_differenced_values,lags=40)
plt.title("ACF (Autocorrelation Function)")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()
```

#### From the above graph, q=0

```python
# Plot the PACF (Partial Autocorrelation Function)
plt.figure(figsize=(12, 6))
plot_pacf(train['Close'], lags=40)
plt.title("PACF (Partial Autocorrelation Function)")
plt.xlabel("Lag")
plt.ylabel("Partial Autocorrelation")
plt.show()
```

#### From the above graph, p=2

#### Choosing s=3 as the cycle repeats every 39 weeks.

```python
df_log = np.log(train["Close"])
df_log
```

```python
# a = auto_arima(df_log,
#                start_p=0,
#                start_q=0,
#                max_p=4,
#                max_q=4,
#                test='adf',
#                seasonal=True,
#                stepwise=True,
#                suppress_warnings=True,
#                d=None,
#                m=52)
# print(a.summary())
```

```python
from statsmodels.tsa.arima.model import ARIMA


p=2
d=1
q=0
s=52
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(df_log, order=(p, d, q), seasonal_order=(1, 1, 0, s), exog=train[["Open","Volume"]])
model_fit_1 = model.fit()

# model_2 = ARIMA(train_series_2, order=(p, d, q))
# model_fit_2 = model_2.fit()
```

```python
preds = model_fit_1.forecast(100, alpha=0.05,exog=test[["Open","Volume"]])
preds = np.exp(preds)
preds
```

```python
t1 = train
t2 = test
t2
```

### Performing one hot encoding for "Strategy"

```python
t1_dummies = pd.get_dummies(t1,columns=["Strategy"]).astype(float)
t1_dummies
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = t1_dummies[["Open","Close","Volume"]]
Y = t1_dummies[["Strategy_Buy","Strategy_Hold","Strategy_Sell"]]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train a decision tree classifier
classifier = DecisionTreeClassifier(max_depth=20,
                                    min_samples_split=2,  # Minimum samples required to split an internal node
                                    min_samples_leaf=4,
                                    criterion='gini',
                                    random_state=42)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(t2[["Open","Close","Volume"]])
```

```python
ans=[]
for i in range(len(y_pred)):
    if y_pred[i][0]==1:
        ans.append("Buy")
    elif y_pred[i][1]==1:
        ans.append("Hold")
    else:
        ans.append("Sell")
ans.count("Buy"), ans.count("Hold"), ans.count("Sell")
```

```python
y_pred=ans
```
